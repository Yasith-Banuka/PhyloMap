```python
Entrez.email = "yoshani.ranaweera.17@cse.mrt.ac.lk"  

for df in dfList:

  !mkdir /content/drive/MyDrive/BioS_Project/Fasta_Files/{df.name}
  print("------------------------> ",df.name)
  accessionList = df['accession'].tolist()
  print(accessionList)

  for accession in accessionList:

    if accession == None:
      continue

    filename = "/content/drive/MyDrive/BioS_Project/Fasta_Files/"+df.name+"/"+accession+".fasta"
    if not os.path.isfile(filename):
        try:
          net_handle = Entrez.efetch(
              db="nucleotide", id=accession, rettype="fasta", retmode="text"
          )
          out_handle = open(filename, "w")
          out_handle.write(net_handle.read())
          out_handle.close()
          net_handle.close()
        except:
          print(accession, "not found in NCBI") 
```
